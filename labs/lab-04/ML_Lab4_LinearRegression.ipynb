#### e. Gradient Descent Algorithm

```
from tqdm import trange
def gradient_descent(alpha, x, y, max_iter=1500):
  theta = 0
  x_trans = x.transpose()
  for i in trange(max_iter):
    loss = np.dot(x, theta) - y
    sum_squared_error = np.square(loss).sum()
    cost = sum_squared_error / (2*len(x))
    # gradient is a function of loss dot X
    gradients = np.dot(x_trans, loss).sum() / len(x)
    theta = theta - alpha * gradients 
  return theta
```

```
theta = gradient_descent(0.01, X_train, y_train, 500)
print(theta)
```

#### f) Try Varying Learning Rates


```
theta_array = []
learning_rates = np.arange(0.1, 1., 0.1)
for rate in learning_rates:
  theta_array.append(
      gradient_descent(rate, X_train, y_train, 10)
  )

plt.plot(learning_rates, theta_array, 'r--')
plt.xlabel('Learning Rates')
plt.ylabel('Theta')
plt.show()
```

## h) Simple K-Fold with error Metrics


```
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import RepeatedKFold
from sklearn.metrics import max_error, mean_absolute_error, mean_squared_error
```

```
skf = RepeatedKFold(n_splits=5, n_repeats=1, random_state=None)
regressor = LinearRegression() 

#keep track of all the performance vectors
from collections import defaultdict
performance = defaultdict(list)
# X is the feature set and y is the target
for train_index, test_index in skf.split(X):
     #print("Train:", train_index, "Validation:", test_index)
     X_train, X_test = X[train_index], X[test_index]
     y_train, y_test = Y[train_index], Y[test_index]
     regressor.fit(X_train, y_train)
     y_pred = regressor.predict(X_test)

     performance['ME'].append(max_error(y_test, y_pred))
     performance['MAE'].append(mean_absolute_error(y_test, y_pred))
     performance['MSE'].append(mean_squared_error(y_test, y_pred))
     performance['RMSE'].append(mean_squared_error(y_test, y_pred, squared=False))


```

```
fig, axs = plt.subplots(2,2, figsize=(20,10))
fig.suptitle('Performances')
 
(me, mae), (mse, rmse) = axs
me.set_title('Max Error')
me.plot(range(5), performance['ME'], 'tab:blue')
mae.set_title('Mean Absolute Error')
mae.plot(range(5), performance['MAE'], 'tab:green')
mse.set_title('Mean Squared Error')
mse.plot(range(5), performance['MSE'], 'tab:orange')
rmse.set_title('Root Mean Squared Error')
rmse.plot(range(5), performance['RMSE'], 'tab:red')
for ax in axs.flat:
  ax.set(xlabel='fold', ylabel='error')
  ax.label_outer()
fig.show()
```

```
print("Max Error:\t\t", np.average(performance['ME']))
print("Max Absolute Error:\t\t", np.average(performance['MAE']))
print("Mean Squared Error:\t\t", np.average(performance['MSE']))
print("Root Mean Square Error:\t\t", np.average(performance['RMSE']))
```

```
from tqdm import trange
def gradient_descent(alpha, x, y, theta, max_iter=1500):
  x_trans = x.transpose()
  for i in trange(max_iter):
    #loss = np.sum(np.dot(x, theta), axis=1) - y
    loss = np.sum(hypothesis(theta, x))
    #sum_squared_error = np.square(loss).sum()
    #cost = sum_squared_error / (2*len(x))
    # gradient is a function of loss dot X
    for c in range(0, len(x[0])):
      #gradient = sum((loss - y) * list(zip(*x))[c]) / len(x)
      gradient = sum((loss - y) * x_trans[c]) / len(x)
      
      #print(c, gradient, alpha)
      theta[c] = theta[c] - alpha * gradient
    #gradients = np.dot(x_trans, loss).sum() / len(x)
    #theta = theta - alpha * gradients 
  return theta
```
